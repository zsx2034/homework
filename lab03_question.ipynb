# Question 1 (9 marks)
Prandtl's lifting-line theory is a mathematical model that predicts the total lift generated by a wing of a finite size. This theory was developed in the early 20th century and it is still in use today for the preliminary calculation of finite-wing characteristics. 

The lifting-line theory states that the lift generated by a finite wing can be calculated by integrating the circulation distribution along  the wing, 


\begin{equation}
L = \int_{-b/2}^{b/2} \rho_\infty V_\infty \Gamma(y) dy
\tag{1}
\end{equation}

where $L$ is the total lift, $b$ is the wing span, $y$ is the distance from the wing root, $\rho_\infty$ is the air density, $V_\infty$ is the air speed, and $\Gamma_0$ is the circulation at the wing root. 


 

The figure above shows the circulation distribution (Image source:  Anderson, J. D. (2007). Fundamentals of aerodynamics. McGraw-Hill.)
