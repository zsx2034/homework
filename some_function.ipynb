```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 排序
x = np.array([1, 3, 2, 5, 4])

# 从小到大排序
sorted_x = np.sort(x)
print("从小到大排序")
print(sorted_x)

# 从大到小排序
sorted_x = np.sort(x)[::-1]
print("从大到小排序")
print(sorted_x)

x = np.array([[1, 4, 2], [5, 3, 6]])
# 按列排序
sorted_x = np.sort(x, axis=0)
print("按列排序")
print(sorted_x)

# 按行排序
sorted_x = np.sort(x, axis=1)
print("按行排序")
print(sorted_x)
```

```python

# 生成等差数列

# 0到10之间生成5个数
tmp = np.linspace(0, 10, 5) 
print("0-10之间生成5个数")
print(tmp)

# 间隔为0.01，也就是0 - 0.49 之间生成50个数
# 包含0和0.49
tmp = np.linspace(0, 0.49, 50)
print("间隔为0.01")
print(tmp)
```

```python

import matplotlib.pyplot as plt

# 定义x和y轴的数据
x = np.linspace(0, 10, 101)
y = np.linspace(0, 10, 101)
x, y = np.meshgrid(x, y)
z = x * y

# 使用scatter函数绘制散点图
# 不建议设置edgecolors
# plt.scatter(x, y, c=z, s=10, edgecolors='black')  
plt.scatter(x, y, c=z, s=10)  # c参数用于设置点的颜色，edgecolors设置边缘颜色，s设置点的大小

# 添加标题和坐标轴标签
plt.title('example')
plt.xlabel('X')
plt.ylabel('Y')

# 显示图形
plt.show()

```
